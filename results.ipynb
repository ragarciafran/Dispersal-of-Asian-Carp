```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import geopandas as gpd
```

```python
species = "bighead"
results = pd.read_excel(f"output/results_{species}.xlsx", index_col=0, na_filter=False)
dike_dict = {"":"None", 0:"D/S", 1:"D/O", 2:"U/S", 3:"U/O"}
seasons = ["spring", "summer", "fall", "winter"]

for i in range(len(results)):
    fig, ax = plt.subplots()
    results_list = results.iloc[i].tolist()
    x = np.arange(1,len(results_list[4:])+1)
    y = results_list[4:]
    ax.plot(x, np.cumsum(y)*10)
    ax.set_title(f'Species: {results_list[0]}, Season: {seasons[results_list[2]]},\nDike: {dike_dict[results_list[3]]}, % entrained: {np.round(np.sum(y)/results_list[1]*100)}%')
    ax.set_xlabel("timestep (hr)")
    ax.set_ylabel("# carp entrained")
    fig.savefig(f'output/{species}/simulation {i}.jpg')
```
